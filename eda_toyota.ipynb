```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/cars.csv")
make = "toyota"
selected_models = ['tacoma', 'camry', 'corolla', 'rav4', 'prius', '4runner']
make_df = df[(df["make"] == make) & (df["model"].isin(selected_models))].copy()
make_df["model"] = pd.Categorical(make_df["model"], categories=selected_models, ordered=True)
make_df = make_df.sort_values("model")
```

### Toyota price outlier cleanup

The selected toyota model subset still has upper-tail price outliers. This step computes the 99th percentile price within that filtered subset and creates `df_clean` by keeping only rows at or below that cutoff. The scatterplots below use `df_clean` so the main price relationships are easier to inspect.


```python
price_upper = make_df["price"].quantile(0.99)

df_clean = make_df[make_df["price"] <= price_upper].copy()

print(f"99th percentile price cutoff: ${price_upper:,.0f}")
print(f"Rows before cleanup: {len(make_df):,}")
print(f"Rows after cleanup: {len(df_clean):,}")
print(f"Rows removed: {len(make_df) - len(df_clean):,}")
```

### Cleaned Toyota price vs odometer

This scatterplot shows price versus odometer reading for the selected representative toyota models after removing the top 1% of prices from the filtered subset. Using `df_clean` reduces the influence of extreme listings while preserving the model comparison.


```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_clean, x="odometer", y="price", hue="model", alpha=0.35)
plt.title("Cleaned Toyota Price vs Odometer")
plt.xlabel("Odometer")
plt.ylabel("Price")
plt.tight_layout()
```

### Cleaned Toyota price vs age

This scatterplot shows price versus age for the selected representative toyota models after removing the top 1% of prices from the filtered subset. The cleaned view makes depreciation patterns easier to inspect across the chosen models.


```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_clean, x="age", y="price", hue="model", alpha=0.35)
plt.title("Cleaned Toyota Price vs Age")
plt.xlabel("Age")
plt.ylabel("Price")
plt.tight_layout()
```

### Median Toyota price by model

This bar chart compares median cleaned price across all toyota models in the dataset, not just the representative subset used in the scatterplots. It first removes the top 1% of prices within the full toyota subset, then summarizes the remaining listings by model.


```python
all_make_df = df[df["make"] == "toyota"].copy()
all_make_price_upper = all_make_df["price"].quantile(0.99)
all_make_df_clean = all_make_df[all_make_df["price"] <= all_make_price_upper].copy()

model_price = (
    all_make_df_clean.groupby("model", as_index=False)["price"]
    .median()
    .sort_values("price", ascending=False)
)

plt.figure(figsize=(14, 6))
sns.barplot(data=model_price, x="model", y="price")
plt.title("Median Toyota Price by Model")
plt.xlabel("Model")
plt.ylabel("Median Price")
plt.xticks(rotation=45, ha="right")
plt.show()
```

### Median Toyota price by condition

This bar chart compares median cleaned price across condition categories for all toyota listings. It uses the full toyota subset after the top 1% price cleanup, so the category comparison is less affected by extreme high-price listings.


```python
condition_price = (
    all_make_df_clean.dropna(subset=["condition"])
    .groupby("condition", as_index=False)["price"]
    .median()
    .sort_values("price", ascending=False)
)

plt.figure(figsize=(10, 5))
sns.barplot(data=condition_price, x="condition", y="price")
plt.title("Median Toyota Price by Condition")
plt.xlabel("Condition")
plt.ylabel("Median Price")
plt.xticks(rotation=45, ha="right")
plt.show()
```
